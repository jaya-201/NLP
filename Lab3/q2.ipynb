```python
import pyarrow.parquet as pq
import matplotlib.pyplot as plt
import os

# Load your tokenized dataset (Parquet format)

table = pq.read_table("tokenized_hi.parquet")
df = table.to_pandas()

# Step 1: Build frequency dictionary manually
word_freq = {}

for row in df["sentences"]:
    for sent in row:
        for token in sent["tokens"]:
            token = token.strip()
            if not token:
                continue
            if token in word_freq:
                word_freq[token] += 1
            else:
                word_freq[token] = 1

print(f"Total unique words: {len(word_freq)}")
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm

# Step 2: Sort and get top 100
sorted_items = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:100]
words = [w for w, c in sorted_items]
counts = [c for w, c in sorted_items]

# Step 3: Font settings (important for Hindi text)
font_path =r"C:\Users\Jaya\Downloads\Noto_Sans_Devanagari\NotoSansDevanagari-VariableFont_wdth,wght.ttf"
prop = fm.FontProperties(fname=font_path)


# Step 4: Plot
plt.figure(figsize=(15,6))
plt.bar(words, counts)
plt.title("Top 100 Most Frequent Hindi Words", fontproperties=prop, fontsize=16)
plt.xticks(rotation=90, fontproperties=prop, fontsize=9)
plt.ylabel("Frequency", fontproperties=prop)
plt.tight_layout()
plt.show()
```

```python
def remove_stopwords(word_freq, threshold):
    return {w: c for w, c in word_freq.items() if c <= threshold}

def plot_top_words(word_freq, threshold, prop, top_n=100):
    sorted_items = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:top_n]
    words = [w for w, c in sorted_items]
    counts = [c for w, c in sorted_items]
    
    plt.figure(figsize=(15,6))
    plt.bar(words, counts)
    
    plt.title(f"Top {top_n} Words (After removing stopwords > {threshold})", 
              fontproperties=prop, fontsize=16)
    plt.xticks(rotation=90, fontproperties=prop, fontsize=9)
    plt.ylabel("Frequency", fontproperties=prop)
    
    plt.tight_layout()
    plt.show()

t=30000
filtered_freq = remove_stopwords(word_freq, t)
plot_top_words(filtered_freq, t, prop)
```
